```python
import pandas as pd

accepts = pd.read_excel("chefmozaccepts.xlsx")
info = pd.read_excel("geoplaces2.xlsx")
parking = pd.read_excel("chefmozparking.xlsx")
cuisine = pd.read_excel("chefmozcuisine.xlsx")
rating = pd.read_excel("rating_final.xlsx")
user = pd.read_excel("userprofile.xlsx")
userpayment = pd.read_excel("userpayment.xlsx")
usercus = pd.read_excel("usercuisine.xlsx")

data_merge = info.merge(accepts,how = "inner",on = "placeID")

payments = data_merge[["placeID","Rpayment","name"]]

payments
```

```python

data_merge = info.merge(parking,how = "inner",on = "placeID")

parking_statuses = data_merge[["placeID","parking_lot","name"]]

parking_statuses
```

```python
data_merge = info.merge(cuisine,how = "inner",on = "placeID")

cuisines = data_merge[["placeID","Rcuisine","name"]]

cuisines
```

```python

data_merge_merge = cuisines.merge(rating,how = "inner",on = "placeID")


for x in data_merge_merge["Rcuisine"].unique():

    cuisine_info = data_merge_merge[data_merge_merge["Rcuisine"] == x]
    print(x)
    print(cuisine_info.describe())
```
